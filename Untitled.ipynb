```python
from skimage import io, color, exposure, img_as_ubyte
import matplotlib.pyplot as plt
import torch
from PIL import Image
import numpy as np
from RealESRGAN import RealESRGAN
import time

# Read the image from file
start_time = time.time()
im1 = io.imread("C:/Users/nipan/OneDrive/Desktop/IMAGE COLORIZATION ML/ll4.jpg")
im1 = color.rgb2gray(im1)
read_time = time.time() - start_time
print(f"Time taken for reading the image: {read_time:.2f} seconds")


# Perform adaptive histogram equalization (AHE) for local brightness enhancement
start_time = time.time()
im1_enhanced = exposure.equalize_adapthist(im1, clip_limit=0.03)
enhancement_time = time.time() - start_time
print(f"Time taken for histogram equalization: {enhancement_time:.2f} seconds")

# Convert the enhanced image to unsigned byte format
start_time = time.time()
im1_enhanced = img_as_ubyte(im1_enhanced)
conversion_time2 = time.time() - start_time
print(f"Time taken for conversion to unsigned byte format: {conversion_time2:.2f} seconds")

# Save the enhanced image
start_time = time.time()
io.imsave('ll3_gray_enhance.png', im1_enhanced)
save_time = time.time() - start_time
print(f"Time taken for saving the enhanced image: {save_time:.2f} seconds")

# image colorization
%run colorization/demo_release.py -i "ll3_gray_enhance.png"

# Load the RealESRGAN model
start_time = time.time()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = RealESRGAN(device, scale=4)
model.load_weights('weights/RealESRGAN_x4.pth', download=True)
load_time = time.time() - start_time
print(f"Time taken for loading the RealESRGAN model: {load_time:.2f} seconds")

# Load the image to be super-resolved
start_time = time.time()
path_to_image = 'saved_eccv16.png'
image = Image.open(path_to_image).convert('RGB')
image_load_time = time.time() - start_time
print(f"Time taken for loading the image to be super-resolved: {image_load_time:.2f} seconds")

# Super-resolve the image
start_time = time.time()
sr_image = model.predict(image)
super_resolution_time = time.time() - start_time
print(f"Time taken for super-resolution: {super_resolution_time:.2f} seconds")


# Display images side by side
fig, axes = plt.subplots(1, 4, figsize=(15, 10))

# Display original grayscale image
axes[0].imshow(im1, cmap='gray')
axes[0].set_title('Original Grayscale Image')

# Display enhanced image
axes[1].imshow(im1_enhanced, cmap='gray')
axes[1].set_title('Enhanced Grayscale Image')

# Display colored image
axes[2].imshow(image)
axes[2].set_title('Colored Grayscale Image')

# Convert the PIL image to a NumPy array for super-resolved image
sr_image_array = np.array(sr_image)


# Display super-resolved image
axes[3].imshow(sr_image_array)
axes[3].set_title('Super-resolved Final Image')


# Hide axis for all subplots
for ax in axes:
    ax.axis('off')

    
plt.tight_layout()
plt.show()
```
